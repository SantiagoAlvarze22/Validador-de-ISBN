```python
def validate_isbn(isbn, length):

    if len(isbn)!= length:
        print(f'ISBN-{length} code should be {length} digits long.')
        return

    main_digits = isbn[0:length-1]

    given_check_digit = isbn[length-1]
    print(type(given_check_digit))

    main_digits_list = [int(digit) for digit in main_digits]

    # Calculate the check digit from other digits
    if length == 10:
        expected_check_digit = calculate_check_digit_10(main_digits_list)
    else:
        expected_check_digit = calculate_check_digit_13(main_digits_list)

    # Check if the given check digit matches with the calculated check digit
    if str(given_check_digit) == expected_check_digit:
        print('Valid ISBN Code.')
    else:
        print('Invalid ISBN Code.')

def calculate_check_digit_10(main_digits_list):
    # Note: You don't have to fully understand the logic in this function.
    digits_sum = 0
    # Multiply each of the first 9 digits by its corresponding weight (10 to 2) and sum up the results
    for index, digit in enumerate(main_digits_list):
        digits_sum += digit * (10 - index)
    # Find the remainder of dividing the sum by 11, then subtract it from 11
    result = 11 - digits_sum % 11
    # The calculation result can range from 1 to 11.
    # If the result is 11, use 0.
    # If the result is 10, use upper case X.
    # Use the value as it is for other numbers.
    if result == 11:
        expected_check_digit = '0'
    elif result == 10:
        expected_check_digit = 'X'
    else:
        expected_check_digit = str(result)
    return expected_check_digit
def calculate_check_digit_13(main_digits_list):
    # Note: You don't have to fully understand the logic in this function.
    digits_sum = 0
    # Multiply each of the first 12 digits by 1 and 3 alternately (starting with 1), and sum up the results
    for index, digit in enumerate(main_digits_list):
        if index % 2 == 0:
            digits_sum += digit * 1
        else:
            digits_sum += digit * 3
    # Find the remainder of dividing the sum by 10, then subtract it from 10
    result = 10 - digits_sum % 10
    # The calculation result can range from 1 to 10.
    # If the result is 10, use 0.
    # Use the value as it is for other numbers.
    if result == 10:
        expected_check_digit = '0'
    else:
        expected_check_digit = str(result)
    return expected_check_digit
def main():
    user_input = input('Enter ISBN and length: ')
    values = user_input.split(',')

    # try:
    #     isbn = int(values[0]) #Valores pasados sean solo enteros sino es asi, tira la excepcion
    # except ValueError:
    #     print('Invalid character was found.')

    try:
        values[1] #Si existe solo un elemento,lanza la excepcion porque no se puede ingresar
    except IndexError:
        print('Enter comma-separated values.')

    try:
        length = int(values[1]) #Si es diferente a un entero
    except ValueError:
        print('Length must be a number.')

    if length == 10 or length == 13:

        if length == 10:

            try:
                convertir = ",".join(values[0]).split(',')

                if convertir[len(convertir) - 1 ] == 'X':

                    lista = list(map(lambda x: int(x), convertir[0:len(convertir) -1 ]))

                    # print(lista)
                    nuev_valor = 'X'
                    el_lista = lista.append(nuev_valor)

                    # print(lista)
                    conv = validate_isbn(lista, length)
                    # print(conv)
                    return conv
                else:
                    validate_isbn(list(map(lambda x: int(x), convertir)), length)

            except ValueError:

                print('Invalid character was found.')

        else:
            try:
                convertir = ",".join(values[0]).split(',')
                validate_isbn(list(map(lambda x: int(x), convertir)), length)
            except ValueError:
                print('Invalid character was found.')
    else:
        print('Length should be 10 or 13.')

main()
```
